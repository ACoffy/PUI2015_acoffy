```python
#Question 1

#Using arrays instead of dictionaries for clearer understanding

#Module Import
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit, minimize

%pylab inline
```

```python
#Random seed to ensure reproducibility
np.random.seed(185) #seed was chosen so that 
```

```python
#Group of 100 unique random integers, between 10 & 2000.
uniqseries = np.random.choice((range(10,2000)),100, replace=False)

#len(genseries) == len(set(genseries)) #check that each value is unique
```

```python
#Mean chosen arbitrarily
genmean = 19
```

```python
#Each distribution is given its own array for mean and sample size

chisqsizeList = []
chisqmeanList = []

binomsizeList = []
binommeanList = []

normalsizeList = []
normalmeanList = []

poissonsizeList = []
poissonmeanList = []

gammasizeList = []
gammameanList = []

logisticsizeList = []
logisticmeanList = []
```

```python
#Loop through & create a random set for each mathematical distribution, with a size
#determined by each value in the random integer group.

for t in uniqseries:
    
    #chisquared distribution sample
    chisqseries = np.random.chisquare(genmean,t) #create the sample of random values
    
    chisqsizeList.append(t) # adds this sample's size to a list of sizes for plotting later
    
    chisqmean = chisqseries.mean() #calculate the mean of the distribution
    chisqmeanList.append(chisqmean) #adds the mean for this series to a list of means
    
    #binomial distribution sample
    binomseries = np.random.binomial(genmean/0.5, 0.5,t) #dividing the genmean in "n" by 0.5 ensures that the n*p yields the set genmean
    binomsizeList.append(t)
    binommean = binomseries.mean()
    binommeanList.append(binommean)
    
    #normal (gaussian) distribution sample
    normseries = np.random.normal(genmean,1,t) 
    normalsizeList.append(t)
    normalmean = normseries.mean() 
    normalmeanList.append(normalmean)
    
    #Poisson distribution sample
    poissonseries = np.random.poisson(genmean, t)
    poissonsizeList.append(t)
    poissonmean = poissonseries.mean()
    poissonmeanList.append(poissonmean)
    
    #Gamma distribution sample
    gammaseries = np.random.gamma(genmean,1,t)
    gammasizeList.append(t)
    gammamean = gammaseries.mean()
    gammameanList.append(gammamean)
    
    #Logistic distribution sample
    logisticseries = np.random.logistic(genmean,size = t)
    logisticsizeList.append(t)
    logisticmean = logisticseries.mean()
    logisticmeanList.append(logisticmean)
```

```python
#Pretty straight forward plotting of each distribution
plt.figure(figsize=(8, 20), dpi=140)      

subplot(321)
chisqplot = plt.scatter(chisqmeanList, chisqsizeList)
plt.ylim(0,2000)
plt.xlabel('Mean')
plt.ylabel('Sample Size')
plt.title('Chi Squared Distribution')
plt.plot([genmean, genmean], [0, 2000], 'k', lw=2) #creates the straight line for the mean

subplot(322)
binomplot = plt.scatter(binommeanList, binomsizeList)
plt.xlim(18, 20)
plt.ylim(0,2000)
plt.xlabel('Mean')
plt.ylabel('Sample Size')
plt.title('Binomial Distribution')
plt.plot([genmean, genmean], [0, 2000], 'k', lw=2)

subplot(323)
normalplot = plt.scatter(normalmeanList, normalsizeList)
plt.xlim(18.5, 19.5)
plt.ylim(0,2000)
plt.xlabel('Mean')
plt.ylabel('Sample Size')
plt.title('Normal Distribution')
plt.plot([genmean, genmean], [0, 2000], 'k', lw=2)

subplot(324)
poissonplot = plt.scatter(poissonmeanList, poissonsizeList)
plt.xlim(18, 20)
plt.ylim(0,2000)
plt.xlabel('Mean')
plt.ylabel('Sample Size')
plt.title('Poisson Distribution')
plt.plot([genmean, genmean], [0, 2000], 'k', lw=2)

subplot(325)
gammaplot = plt.scatter(gammameanList, gammasizeList)
plt.xlim(18, 20)
plt.ylim(0,2000)
plt.xlabel('Mean')
plt.ylabel('Sample Size')
plt.title('Gamma Distribution')
plt.plot([genmean, genmean], [0, 2000], 'k', lw=2)

subplot(326)
logisticplot = plt.scatter(logisticmeanList, logisticsizeList)
#plt.xlim(18, 20)
plt.ylim(0,2000)
plt.xlabel('Mean')
plt.ylabel('Sample Size')
plt.title('Logistic Distribution')
plt.plot([genmean, genmean], [0, 2000], 'k', lw=2)
```

Most of the graphs look as they should particularly the normal distribution. Very tight and clustered around the mean. The poisson doesn't immediately look strikingly like it should but a histogram plot confirms that the distribution is correct. The scatter plot doesn't clearly convey the left tendency expected in the poisson but it can be seen slightly.

```python
dist = 6
bins = 20
alpha = .2

plt.figure(figsize=(16, 8), dpi=140) 
plt.ylim(0,15)
plt.hist(chisqmeanList, bins=20, histtype='stepfilled', normed=True, color='yellow', alpha=.8, label='Chi Squared')
plt.hist(binommeanList, bins=20, histtype='stepfilled', normed=True, color='green', alpha=.6, label='Binomal')
plt.hist(normalmeanList, bins=20, histtype='stepfilled', normed=True, color='violet', alpha = .5, label='Normal')
plt.hist(poissonmeanList, bins=20, histtype='stepfilled', normed=True, color='blue', alpha=.5 , label='Poisson')
plt.hist(gammameanList, bins=20, histtype='stepfilled', normed=True, color='pink', alpha=.8, label='Gamma')
plt.hist(logisticmeanList , bins=20, histtype='stepfilled', normed=True, color='purple', alpha=.6, label='Logistic')

plt.title("Historgram of Distributions Sample Means")
plt.xlabel("Mean")
plt.ylabel("Number of Occurances")
plt.legend()
plt.show()
```
